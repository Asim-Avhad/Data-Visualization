```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
import kagglehub

path = kagglehub.dataset_download("nitindatta/finance-data")
data = pd.read_csv(path + '/Finance_data.csv')
data.dropna(inplace=True)
```

```python
kmeans = KMeans(n_clusters=3)
data['Cluster_Investments'] = kmeans.fit_predict(data[['Mutual_Funds', 'Equity_Market', 'Fixed_Deposits', 'Gold']])

# Plotting the clusters based on investment preferences
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Mutual_Funds', y='Equity_Market', hue='Cluster_Investments', data=data, palette='Set1')
plt.title('Customer Segmentation Based on Investment Preferences')
plt.xlabel('Mutual Funds')
plt.ylabel('Equity Market')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data['age'].hist(bins=30)
plt.title('Age Distribution of Investors')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
columns_to_convert = ['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold', 'Stock_Marktet']
data[columns_to_convert] = data[columns_to_convert].apply(pd.to_numeric, errors='coerce')

plt.figure(figsize=(12, 8))
correlation = data[columns_to_convert].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Investment Avenues')
plt.show()
```
